```python
import pandas as pd
import os
os.chdir("../mysql")
from mysqlfunctions import load_my_sql, full_stat_extract
os.chdir("../misc. data pulls")
```

```python
connection = load_my_sql('aik.9Yor')
all_diarynums = connection.execute('SELECT diary FROM notes;')



def get_final_ids():
    dir_path = "../final/data/"
    directory = os.path.join(dir_path)
    
    # used for step 4/5
    diary_id_ls = []
    # step 1
    for root,dirs,files in os.walk(directory):
        for file in files:
            # step 2
            if file.endswith(".csv"):
                filepath = dir_path + file
                f=pd.read_csv(filepath, encoding = 'latin1')
                diary_id_ls = diary_id_ls + [int(i) for i in f['diarynum']]
    return list(set(diary_id_ls))
            
```

```python
final_ids = get_final_ids()
```

```python
def create_result_df(connection, final_ids):
    results = []

    for i in final_ids:
        results.append(full_stat_extract(connection, i)[0])
    print(results)
    df = pd.DataFrame(results, columns = ['firstName', 'lastName', 'info', 'gender', 'birthDay', 'pub']) 
    return df
```

```python
full_info = create_result_df(connection, final_ids)
```
